```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import PolyCollection
```

```python
points = np.loadtxt('points.txt')
quads = np.loadtxt('quads.txt', dtype=int)

# Build list of quad polygons
polygons = []

for quad in quads:
    polygon = points[quad]
    polygons.append(polygon)

fig, ax = plt.subplots()
ax.set_aspect('equal')

pc = PolyCollection(polygons,
                    edgecolor='black',
                    facecolor='lightblue')

ax.add_collection(pc)
ax.autoscale()

plt.title("Quadrilateral Mesh")
plt.show()
```
